```python
import re
import pandas as pd


df = pd.read_csv('C:\\Users\\ryan\\Desktop\\Thinkful DS Sample Data - Main Course\\WELLCOME_APCspend2013_forThinkful3.csv')
df.head()
```

```python
#Explore the data with a groupby to help visualize some of the problems
df2 = df.groupby('Journal title').count()
df2.head(1000)
#print(df2)
#Will need to clear spaces before and after, remove commas and semicolons, capitalize all titles
```

```python
#Remove blank spaces before and after Journal title
df['Journal Title 2'] = df['Journal title'].apply(lambda x: str(x).strip())
#Remove commas and colons
df['Journal Title 3'] = df['Journal Title 2'].apply(lambda x: str(x).replace(':', ''))
df['Journal Title 4'] = df['Journal Title 3'].apply(lambda x: str(x).replace(',', ''))
#Put everything in all caps
df['Journal Title 5'] = df['Journal Title 4'].apply(lambda x: str(x).upper())
#Remove all text after " Section" to better group journals together
df['Journal Title 6'] = df['Journal Title 5'].apply(lambda x: str(x).split(' SECTION', 1)[0])
#Replace all " J " with "Journal"
df['Journal Title 7'] = df['Journal Title 6'].apply(lambda x: str(x).replace('J ', 'JOURNAL '))
df['Journal Title 8'] = df['Journal Title 7'].apply(lambda x: str(x).strip())
 
#Remove commas, $ signs from cost
df['Cost New'] = df['COST () charged to Wellcome (inc VAT when charged)'].apply(lambda x: str(x).replace(',', ''))
df['Cost New 2'] = df['Cost New'].apply(lambda x: str(x).replace('$', ''))
#Convert cost variable from string to a float to allow for mathematical operations
df['Cost New 3'] = df['Cost New 2'].apply(lambda x: float(x))

#print(df['Cost New 3'].max())
#Some cost amounts are 999.9K - exclude these, replace with median
df['Cost New 4'] = df['Cost New 3'].replace(999999.00, df['Cost New 3'].median())
#print(df['Cost New 4'].min())
#Some cost amounts are 0 - exclude these, replace with median
df['Cost New 5'] = df['Cost New 4'].replace(0.00, df['Cost New 3'].median())

#Sort by number of articles published
#df2 = df.groupby('Journal Title 8', as_index = False).count()
#df3 = df2.sort_values('Journal Title 7', ascending = False)
#df3.head()
```

```python
#Calculate mean, median, and mode

df5 = pd.DataFrame()
df5['Journal'] = df['Journal Title 8']
df5['Cost'] = df['Cost New 5']
#df5.head()
#df5[df5['Journal'] == 'JOURNAL MED CHEM']
#df2[df2['Journal Title 8'] == 'JOURNAL MED CHEM']
df_count = df5.groupby('Journal', as_index = False).count()
df_count.head()
```

```python
df_mean = df5.groupby('Journal', as_index = False).mean()
df_mean.head()
```

```python
df_med = df5.groupby('Journal', as_index = False).median()
df_med.head()
```

```python
df_std = df5.groupby('Journal').std(ddof=0)
#df_std.head()
#df5.std()
df_std.head()
```

```python
df_final = pd.DataFrame()
df_final['Journal'] = df_count['Journal']
df_final['Count'] = df_count['Cost']
df_final['Mean'] = df_mean['Cost']
df_final['Med'] = df_med['Cost']
#df_final['StDev'] = df_std[0] -- ********Why can't I use as_index = False on STDEV?*******
df_final.head()
```

```python
import numpy as np
result = df5.groupby('Journal', as_index=False).agg([np.count_nonzero,
                                                       np.mean, 
                                                       np.median, 
                                                       lambda x: np.std(x, ddof=0)])

result.head()
```
