```python
djia['Label'] = np.where(djia['Close'] > djia['Open'], 1, 0)
```

```python
testmerge = pd.merge(news, djia[['Date', 'Label']], on='Date', how = 'left')
```

```python
df_new = pd.DataFrame()
counter = 1
row_index = 0
for row in range(len(testmerge)):
    if counter < 26:
        df_new.iloc[row_index][counter] = testmerge.iloc[row]['News']
        counter = counter + 1
        print('step1row',row)
        if counter >= 26:
            counter = 1
            row_index = row_index + 1
            df_new.iloc[row_index][counter] = testmerge.iloc[row]['News']
```
